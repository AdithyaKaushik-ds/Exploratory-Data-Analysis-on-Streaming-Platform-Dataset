# **Basic Analysis**



##**1.   Un-nesting the columns!**

  *   Un-nest the columns those have cells with multiple comma separated values by creating multiple rows.

```python
df['listed_in'] = df['listed_in'].str.split(', ')
df = df.explode('listed_in')

df['cast'] = df['cast'].str.split(', ')
df = df.explode('cast')

df['country'] = df['country'].str.split(', ')
df = df.explode('country')

df['director'] = df['director'].str.split(', ')
df = df.explode('director')
```

* Insights: By utilizing the .explode() function, I have successfully separated the combined values in the 'cast' and 'listed_in' columns, which were originally comma-separated. This process of un-nesting is crucial for ensuring that each value is individually accessible for analysis, preventing any complications that may arise from having multiple values consolidated within a single cell.





##**2. Handling null values**

*   For categorical variables with null values, update those rows as
  unknown_column_name.
  
  Example : Replace missing value with Unknown Actor for missing value in
  Actors column.
*   Replace with 0 for continuous variables having null values.

```python
# Getting a count of all the NaN/null values in each column
df.isna().sum()
```

```python
# Replacing all the null/Nan vales in Director column with alias as 'Unknown Director'
df['director'].fillna('Unknown Director',inplace = True)

# Replacing all the null/Nan vales in Cast column with alias as 'Unknown Actor'
df['cast'].fillna('Unknown Actor',inplace = True)

# Replacing all the null/Nan vales in Country column with alias as 'Unknown'
df['country'].fillna('Unknown',inplace = True)

# Replacing all the null/Nan vales in duration column with alias as 'Unknown'
df['duration'].fillna('Unknown',inplace = True)

# Replacing all the null/Nan vales in rating column with alias as 'Unknown'
df['rating'].fillna('Unknown',inplace = True)

# Dropping the Null values in the date columns
df = df.loc[~df['date_added'].isnull()]
df = df.loc[~df['release_year'].isnull()]
```

```python
# Checking for Duplicate values
df.duplicated().sum()
```

```python
# Removing the Duplicate Values
df = df.drop_duplicates(keep = 'first')
```

```python
df.isna().sum()
```

```python
df.info()
```

* Insights: I have replaced all NaN/null values with their respective aliases in each column, then identified and eliminated any duplicate values. Finally when I checked the sum of Null/NaN values was zero.




```python
# Converting the 'date_added' column to datetime format

df['date_added'] = pd.to_datetime(df['date_added'])
```

##**3. What is the best time to launch a TV show?**


1.   Find which is the best week to release the Tv-show or the movie. Do the analysis separately for Tv-shows and Movies.

  Hint : We expect you to create a new column and group by each week and count
the total number of movies/ tv shows.

```python
movies = df[df['type'] == 'Movie']
tv_shows = df[df['type'] == 'TV Show']
```

```python
#Grouping movies weekly
weekly_movies = movies.groupby(['week_added'])['title'].nunique().sort_values(ascending = False)
weekly_movies.name = 'count' # to avoid error while doing reset_index
weekly_movies = weekly_movies.reset_index()
weekly_movies.head()
```

```python
weekly_movies['week_added'] = weekly_movies['week_added'].astype('int64')
weekly_movies.info()
```

```python
#Grouping tv shows weekly
weekly_tvshows = tv_shows.groupby(['week_added'])['title'].nunique().sort_values(ascending = False)
weekly_tvshows.name = 'count' # to avoid error while doing reset_index
weekly_tvshows = weekly_tvshows.reset_index()
weekly_tvshows.head()
```

```python
weekly_tvshows['week_added'] = weekly_tvshows['week_added'].astype('int64')
weekly_tvshows.info()
```

```python
#Plotting
#setting the plot style
fig = plt.figure(figsize = (15,7))
plt.suptitle('Best Week to launch a Movie/TV show?')
plt.subplot(2,1,1)
sns.lineplot(data = weekly_movies, x = 'week_added', y = 'count',marker = 'o',markersize = 10,color ='#b20710',
 label = 'Movie')

#customizing axis labels
plt.xlabel('Week Added',fontweight = 'bold',fontfamily='serif')
plt.ylabel('Count',fontweight = 'bold',fontfamily='serif')

# creating graph for count of tv shows
plt.subplot(2,1,2)
sns.lineplot(data = weekly_tvshows, x = 'week_added', y = 'count',marker = 'o',markersize = 10,color ='#221f1f',
 label = 'TV Show')

#customizing axis labels
plt.xlabel('Week Added',fontweight = 'bold',fontfamily='serif')
plt.ylabel('Count',fontweight = 'bold',fontfamily='serif')
plt.show()
```

Insights:
* The weekly upload count of both Movies and TV shows exhibits a remarkably similar trend with respect to spikes and dips.
* A unique pattern can be observed in terms of spikes and dips. During a specific week, there is a noticeable increase in content uploads, followed by a three-week period of reduced content addition. This pattern might indicate a monthly cycle wherein the majority of content is uploaded during the first week and the initial week of the subsequent month.
* The highest volume of movies are typically uploaded during the initial week of the year.
* The highest volume of TV shows tends to be added in the 26th week (around June end)

##**6. Find After how many days the movie will be added to Netflix after the release of the movie (you can consider the recent past data)**

Hint : We want you to get the difference between the columns having date added
information and release year information and get the mode of difference. This
will give an insight into what will be the better time to add in Netflix

```python
import pandas as pd

# Load the Netflix dataset
netflix_data = pd.read_csv('netflix_titles.csv')

# Filter out rows where movie release year and date added are available
release_added_data = netflix_data[['release_year', 'date_added']].dropna()

# Convert date_added column to datetime format
release_added_data['date_added'] = pd.to_datetime(release_added_data['date_added'])

# Calculate time difference (in days) between date added and release year
release_added_data['time_difference'] = (release_added_data['date_added'] - release_added_data['release_year']
                                         .apply(lambda x: pd.to_datetime(str(int(x)), format='%Y'))).dt.days

# Calculate the mode of time differences
mode_time_difference = release_added_data['time_difference'].mode()[0]

print("Mode time difference (in days):", mode_time_difference)
```

Insights:
The most common delay between a movie's release and its addition to Netflix is 334 days. This suggests that Netflix typically adds movies to its platform around 334 days after their release.
